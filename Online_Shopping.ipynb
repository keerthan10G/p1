```python
val=df["Product_Category"].value_counts()
l=[]
for i in val:
    if(i>3000):
       l.append(i)
ser=pd.Series(l)
ser
```

```python
from  matplotlib.gridspec import GridSpec

pl.style.context('seaborn-v0_8.bright')
gs=GridSpec(2,2)
fig=pl.figure(figsize=(6,6),dpi=200)
fig1=pl.figure(figsize=(5,6),dpi=200)
fig2=pl.figure(figsize=(5,4),dpi=200)

ax1=fig.add_subplot(gs[0,0])
l=[]
for i in range(0,5):
    l.append(0.05)
ax1.set_title("Most Selled products",fontsize=15,fontstyle="oblique",color='#d62727',fontweight='bold')
ax1.pie(ser,labels=["Apparel","Nest-USA","Office","Drinkware","Lifestyle"],explode=l,autopct="%.1f%%",textprops={'font':'arial','fontsize':10},pctdistance=0.7)
fig.tight_layout()
fig.savefig("mostsold.png",dpi=300,bbox_inches='tight')

ax2=fig1.add_subplot(gs[1,1])
ax2.bar(df[df["Product_Category"]=="Apparel"]["Product_Category"],df[df["Product_Category"]=="Apparel"]["Discount_pct"].mean(),color="#4287f5",width=0.5)
ax2.bar(df[df["Product_Category"]=="Nest-USA"]["Product_Category"],df[df["Product_Category"]=="Nest-USA"]["Discount_pct"].mean(),color="#42f5e9",width=0.5)
ax2.bar(df[df["Product_Category"]=="Office"]["Product_Category"],df[df["Product_Category"]=="Office"]["Discount_pct"].mean(),color="#e6e207",width=0.5)
ax2.set_title("Average Discounts on Products",fontsize=10,fontstyle="oblique",color='#d62727',fontweight='bold')
ax2.set_xlabel("Top 3 Sold products",color="#a832a6",fontdict={'fontstyle':'italic','fontsize':10})
ax2.set_ylabel("Discount %",color="#a832a6",fontdict={'fontstyle':'italic','fontsize':10})
ax2.set_yticks([0,10,20,30])
fig1.tight_layout()
fig1.savefig("mostsold1.png",dpi=300,bbox_inches='tight')


ax3=fig2.add_subplot(gs[1,0])
ax3.bar(df[df["Product_Category"]=="Apparel"]["Product_Category"],df[df["Product_Category"]=="Apparel"]["Delivery_Charges"].mean(),color="#4287f5",width=0.5)
ax3.bar(df[df["Product_Category"]=="Nest-USA"]["Product_Category"],df[df["Product_Category"]=="Nest-USA"]["Delivery_Charges"].mean(),color="#42f5e9",width=0.5)
ax3.bar(df[df["Product_Category"]=="Office"]["Product_Category"],df[df["Product_Category"]=="Office"]["Delivery_Charges"].mean(),color="#e6e207",width=0.5)
ax3.set_title("Average Delivery charges on Product",fontsize=10,fontstyle="oblique",color='#d62727',fontweight='bold')
ax3.set_xlabel("Top 3 Sold products",color="#a832a6",fontdict={'fontstyle':'italic','fontsize':10})
ax3.set_ylabel("Delivery Charges",color="#a832a6",fontdict={'fontstyle':'italic','fontsize':10})
fig2.tight_layout()
fig2.savefig("mostsold2.png",dpi=300,bbox_inches='tight')

pl.show()

```
